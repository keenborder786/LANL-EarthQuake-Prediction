```python
def preprocess(df):
  # df.fillna(0,inplace=True)
  df=df.replace([np.inf, -np.inf], np.nan).dropna()
  X=np.array(df.drop("time_to_failure",1))
  y=np.array(df["time_to_failure"])
  preprocessing_pipeline = Pipeline(steps=[
      ('scaler', StandardScaler())])
  X_transformed=preprocessing_pipeline.fit_transform(X)
  return X_transformed,y
```

```python
def regression_cv(clf,X,y,cv,ss):
    random_number=np.random.randint(X.shape[0], size=ss)

    X_sample=X[random_number, :]
    y_sample=y[random_number]

    cv_results=cross_validate(clf,X_sample,y_sample,cv=cv)
    sorted(cv_results.keys())                         
    ['fit_time', 'score_time', 'test_score']
    
    print("The test scores for {} fold validation are with sample size".format(cv,ss),cv_results['test_score'])
    print("The fit time for {} fold validation are with sample size".format(cv,ss),cv_results['fit_time'])
    plt.plot(list(range(0,5,1)),cv_results['fit_time'],color="red",label="fit_time")
    plt.plot(list(range(0,5,1)),cv_results['test_score'],color="green",label="testing_score")
    plt.title("Cross validation result for {}".format(str(clf)[:20]))
    plt.xlabel("CVs")
    plt.ylim(0,1)
    plt.ylabel("Score")
    plt.legend()
    plt.show()
```

```python
X,y=preprocess(df)
regression_cv(Lasso(),X,y,5,1500000)
```

```python
def random_sample_testing(clf,ss):
    from tqdm import tqdm
    random_number=np.random.randint(X.shape[0], size=ss)
    X_sample=X[random_number, :]
    y_sample=y[random_number]

    
    trails=[]
    scores=[]
    for i in tqdm(range(100)):
      X_train,X_test,y_train,y_test=train_test_split(X_sample,y_sample)
      clf.fit(X_train,y_train)
      scores.append(clf.score(X_test,y_test))
      trails.append(i)
    figure=plt.figure()
    ax=plt.subplot(111)
    plt.plot(trails,scores)
    ax.set_xlabel("Trials")
    ax.set_ylabel("score")
    ax.set_title("Score over {} for {}".format(len(trails),str(clf[:10])))
  
```

```python
random_sample_testing(GradientBoostingRegressor(),150000)
```

```python
import gc
import xgboost as xgb
incremental_dataframe = pd.read_csv("train_feature.csv",
                        chunksize=150000)

xgb_params = {
  'update':'refresh',
  'refresh_leaf': True,
  'silent': False,'eval_metric':'rmse','verbosity':2}

xgb_estimator = None
for df in incremental_dataframe:
  df.drop(df.columns[0], axis=1,inplace=True)
  X_transformed,y = preprocess(df)

 
  xtrain, xvalid, ytrain, yvalid = train_test_split(X_transformed,y)
  xgb_model = xgb.train(xgb_params, 
                        dtrain=xgb.DMatrix(xtrain, ytrain),
                        evals=[(xgb.DMatrix(xvalid, yvalid),"Valid")],
                        xgb_model = xgb_estimator)
 
  
  del df, xtrain, ytrain, xvalid, yvalid
  gc.collect()
xgb_model.save_model('mainxgboost.model')
```

```python
from keras.models import Sequential
from keras.layers import Dense,Dropout
from sklearn.model_selection import train_test_split
from tqdm import tqdm

model = Sequential()
model.add(Dense(64, input_dim=22, activation= "relu"))
model.add(Dropout(.3))
model.add(Dense(1))


model.compile(loss= "mean_squared_error" , optimizer="adam", metrics=["mean_squared_error"])

incremental_dataframe = pd.read_csv("train_feature.csv",
                        chunksize=150000)
a=1
val_loss_idf=[]
val_mean_squared_error_idf=[]
for df in tqdm(incremental_dataframe):
  df.drop(df.columns[0], axis=1,inplace=True)
  X_transformed,y = preprocess(df)
  X_train, X_valid, y_train, y_valid = train_test_split(X_transformed,y)
  if a!=1:
    keras.models.load_model("my_keras_model.h5")
  history=model.fit(X_train,y_train,epochs=1,validation_data=(X_valid, y_valid))
  val_loss_idf.append(history.history["val_loss"])
  val_mean_squared_error_idf.append(history.history["val_mean_squared_error"])
  model.save("my_keras_model.h5")
  a+=1



```
